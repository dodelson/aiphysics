```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam = np.arange(-2,1,.1)
lam=10.**lam
```

```python
k=1.
s=np.sqrt((k-np.log(lam))*lam/(1.+lam))
plt.plot(lam,s,label='1')
k=2.
s=np.sqrt((k-np.log(lam))*lam/(1.+lam))
plt.plot(lam,s,label='2')
k=10.
s=np.sqrt((k-np.log(lam))*lam/(1.+lam))
plt.plot(lam,s,label='10')
plt.legend()
plt.xscale('log')
plt.xlabel('C_s/C_n')
plt.ylabel('$s/C_n^{1/2}$')
plt.title('Constant Log Likelihood Curves')
```

```python
sarr=np.arange(.01,3.,.1)
```

```python
d=2.5
def likelihood(l,s):
    tmp=np.log(l)+(s-d)**2 +s**2/l
    return tmp

n=np.size(lam)
m=np.size(sarr)
img=np.zeros((n,m))
for i in range(n):
    for j in range(m):
        img[i,j]=likelihood(lam[i],sarr[j])
print(img[0,0],img[0,29])
```

```python
plt.imshow(np.exp(-.5*img))
plt.xticks(ticks=[0,10,20,29],labels=[.01,.1,1.,10])
plt.yticks(ticks=[0,10,20,29],labels=[0,1.,2.,3.])
plt.xlabel('C_s/C_n')
plt.ylabel('$s/C_n^{1/2}$')
plt.colorbar()
plt.title('Likelihood $d=2C_n^{1/2}$')
plt.savefig('l2.png')
```

```python
d=2.5
def like1d(l):
    tmp=np.log(l+1.) + d**2/(1.+l)
    return tmp
larr1d=like1d(lam)
```

```python
plt.plot(lam,np.exp(-.5*larr1d))
plt.xlabel('$C_s/C_n$')
plt.ylabel('1D Likelihood ($d=2C_n^{1/2}$)')
plt.savefig('like1d.png')
```

```python
laml=np.arange(.01,10,.3)
n=np.size(laml)
m=np.size(sarr)
img=np.zeros((n,m))
for i in range(n):
    for j in range(m):
        img[i,j]=likelihood(laml[i],sarr[j])
```
