```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#From
#https://www.kaggle.com/lucidlenn/sloan-digital-sky-survey
sdss=pd.read_csv('Skyserver_SQL2_27_2018 6_51_39 PM.csv')
```

```python
sdss.info()
```

```python
## Convert class column to integer
sdss["class"]=np.where(sdss["class"]=='QSO','2',sdss["class"])#.astype(int)
sdss["class"]=np.where(sdss["class"]=='STAR','0',sdss["class"])#.astype(int)
sdss["class"]=np.where(sdss["class"]=='GALAXY','1',sdss["class"])#.astype(int)
sdssc=sdss.astype({'class':int})
sdssc.info()
```

```python
sdssc.describe()
```

```python
starsOnly = sdssc[sdssc["class"]==0]
galaxiesOnly = sdssc[sdssc["class"]==1]
qsoOnly = sdssc[sdssc["class"]==2]
```

```python
starsOnly.describe()
```

```python
galaxiesOnly.describe()
```

```python
qsoOnly.describe()
```

```python
#Visualize the data for all objects
sdssc.plot(kind="scatter",x="ra",y="dec",c="class",alpha=0.1,cmap=plt.get_cmap("jet"),colorbar=True)
```

```python
sdssc.plot(kind="scatter",x='redshift',y='plate')
#This is going to cause problems, because there are really weird, nonlinear correlations that will mess things up
```

```python
galaxiesOnly.plot(kind="scatter",x='redshift',y='plate')
```

```python
qsoOnly.plot(kind="scatter",x='redshift',y='plate')
```

```python
#Get a training set for all objects
X_train_all_full=sdssc[:8000]
y_train_all_label=X_train_all_full["redshift"]
print(X_train_all_full.shape)
print(y_train_all_label.shape)
# Look at correlations
corr_matrix=X_train_all_full.corr()
corr_matrix["redshift"].sort_values(ascending=False)
#corr_matrix["class"].sort_values(ascending=False)
plt.hist(y_train_all_label,bins=50)
```

```python
#Get a training set for the stars
X_train_stars_full=starsOnly[:3300] #80% of stars is ~3300
y_train_stars_label=X_train_stars_full["redshift"]
print(X_train_stars_full.shape)
print(y_train_stars_label.shape)
# Look at correlations
corr_matrix_stars=X_train_stars_full.corr()
corr_matrix_stars["redshift"].sort_values(ascending=False)
#corr_matrix["class"].sort_values(ascending=False)
plt.hist(y_train_stars_label,bins=50)
```

```python
#Get a training set for the galaxies
X_train_galaxies_full=galaxiesOnly[:4000] #80% of stars is ~4000
y_train_galaxies_label=X_train_galaxies_full["redshift"]
print(X_train_galaxies_full.shape)
print(y_train_galaxies_label.shape)
# Look at correlations
corr_matrix_galaxies=X_train_galaxies_full.corr()
corr_matrix_galaxies["redshift"].sort_values(ascending=False)
#corr_matrix["class"].sort_values(ascending=False)
plt.hist(y_train_galaxies_label,bins=50)
```

```python
#Get a training set for the QSOs
X_train_qso_full=qsoOnly[:680] #80% of QSOs is 680
y_train_qso_label=X_train_qso_full["redshift"]
print(X_train_qso_full.shape)
print(y_train_qso_label.shape)
# Look at correlations
corr_matrix_qso=X_train_qso_full.corr()
corr_matrix_qso["redshift"].sort_values(ascending=False)
#corr_matrix["class"].sort_values(ascending=False)
plt.hist(y_train_qso_label,bins=50)
```

```python
# So the stars and galaxies are pretty nicely peaked, while the QSOs are just all over the place.

# Drop all unimportant features.
X_train=X_train_full.drop(columns=["mjd","redshift","plate","fiberid","specobjid","field","camcol","run","rerun","objid","ra","dec"],axis=1)#,inplace=True)
X_train_stars=X_train_stars_full.drop(columns=["mjd","redshift","plate","fiberid","specobjid","field","camcol","run","rerun","objid","ra","dec"],axis=1)#,inplace=True)
X_train_galaxies=X_train_galaxies_full.drop(columns=["mjd","redshift","plate","fiberid","specobjid","field","camcol","run","rerun","objid","ra","dec"],axis=1)#,inplace=True)
X_train_qso=X_train_qso_full.drop(columns=["mjd","redshift","plate","fiberid","specobjid","field","camcol","run","rerun","objid","ra","dec"],axis=1)#,inplace=True)
print(X_train_full.shape,X_train.shape)
X_train.describe()
```

```python
#Let's only focus on galaxies and QSOs here
#Skipping straight to the normalizing step...
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression(normalize=True)
result1_norm=lin_reg.fit(X_train_galaxies,y_train_galaxies_label)
coeff1_norm=result1_norm.coef_
print(coeff1_norm)
print(lin_reg.predict(X_train_galaxies[:2]))
print(y_train_galaxies_label[:2])
plt.scatter(lin_reg.predict(X_train_galaxies),y_train_galaxies_label)
plt.xlabel('$z_{predicted}$')
plt.ylabel('$z_{true}$')
plt.axis([0,1.0,0,1.0])

print('Score= '+str(lin_reg.score(X_train_galaxies,y_train_galaxies_label)))

```

```python
lin_reg=LinearRegression(normalize=True)
result2_norm=lin_reg.fit(X_train_qso,y_train_qso_label)
coeff2_norm=result2_norm.coef_
print(coeff2_norm)
print(lin_reg.predict(X_train_qso[:2]))
print(y_train_qso_label[:2])
plt.scatter(lin_reg.predict(X_train_qso),y_train_qso_label)
plt.xlabel('$z_{predicted}$')
plt.ylabel('$z_{true}$')
plt.axis([0,5.5,0,5.5])

print('Score= '+str(lin_reg.score(X_train_qso,y_train_qso_label)))
```

```python
### This does better but still not well. 
### Look at the coefficients that are largest here are:
ind=np.argsort(np.abs(coeff1_norm))
for i in ind:
    print(X_train_galaxies.columns[i],coeff1_norm[i])
print(X_train_galaxies.columns)

ind=np.argsort(np.abs(coeff2_norm))
for i in ind:
    print(X_train_qso.columns[i],coeff2_norm[i])
print(X_train_qso.columns)
```

```python
# The galaxies look decent enough; focusing on the QSOs...
# With that kind of shape, maybe a polynomial regression would suit it better.
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
degree = 4 
#Rather arbitrary degree, I might have misunderstood the linear regression but an even degree polynomial should match that, and it looks a little too steep to be degree 2.
polyreg = make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X_train_qso,y_train_qso_label)

plt.scatter(polyreg.predict(X_train_qso),y_train_qso_label)
plt.xlabel('$z_{predicted}$')
plt.ylabel('$z_{true}$')
plt.axis([0,5.5,0,5.5])

print('Score= '+str(polyreg.score(X_train_qso,y_train_qso_label)))
# A perfectly reasonable score in my opinion.
# (It should be noted that I have very little idea what's going on anymore)
```
