```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#From
#https://www.kaggle.com/lucidlenn/sloan-digital-sky-survey
sdss=pd.read_csv('C:/Users/subha/downloads/Skyserver_SQL2_27_2018 6_51_39 PM.csv')
```

```python
sdss.info()
```

```python
#filter out only galaxies
galaxies = sdss[sdss["class"] == "GALAXY"]
```

```python
galaxies.info()
```

```python
# only relevant values are u, g, r, i, z - others are noise
# extract redshift and relevant values
X = galaxies[["u", "g", "r", "i", "z"]]
y = galaxies[["redshift"]]
```

```python
#quick scatter plot function
def scatter_plot(feature, target):
    plt.figure(figsize=(16, 8))
    plt.scatter(
        galaxies[feature],
        galaxies[target],
        c='black'
    )
    plt.xlabel("feature")
    plt.ylabel("redshift")
    plt.show()
```

```python
scatter_plot("u", "redshift")
```

```python
scatter_plot("g", "redshift")
```

```python
scatter_plot("r", "redshift")
```

```python
scatter_plot("i", "redshift")
```

```python
scatter_plot("z", "redshift")
```

```python
#all seem to be related to redshift
```

```python
#import needed packages for Ridge regression and cross validation
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import RidgeCV
#import package for polynomial regression (later)
from sklearn.preprocessing import PolynomialFeatures
```

```python
#using variety of params, do cross validation on regularized regression model
params = [0, 1e-15, 1e-10, 1e-7, 1e-5, 1e-3, 1e-2, 1e-1, 1, 10, 100]
#no normalization, won't change mse error but will help get scale of polynomial features
ridge_regressor = RidgeCV(params, cv = 10, scoring = "neg_mean_squared_error")
model = ridge_regressor.fit(X, y)
```

```python
model.score(X, y)
```

```python
#plot residuals graph
plt.figure(figsize=(16, 8))
plt.scatter(
    galaxies["redshift"],
    np.concatenate(model.predict(X)) - galaxies["redshift"],
    c="black"
)
```

```python
#bad score with regularities in residuals
```

```python
#try polynomial regression, regularization parameter along with cv should help with overfitting
poly = PolynomialFeatures(degree = 5)
X_poly = poly.fit_transform(X)
poly_model = ridge_regressor.fit(X_poly, y)
```

```python
poly_model.score(X_poly, y)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    galaxies["redshift"],
    np.concatenate(poly_model.predict(X_poly)) - galaxies["redshift"],
    c="black"
)
```

```python
#much better residuals graph as desired
```
