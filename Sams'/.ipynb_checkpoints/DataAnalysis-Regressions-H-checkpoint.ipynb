```python
housing_data=df.copy()


#DEAL WITH CATEGORICAL VARIABLES
suburbDict={list(set(housing_data["Suburb"]))[k]:k for k in range(len(set(housing_data["Suburb"])))}
typeDict={list(set(housing_data["Type"]))[k]:k for k in range(len(set(housing_data["Type"])))}
methDict={list(set(housing_data["Method"]))[k]:k for k in range(len(set(housing_data["Method"])))}
sellerDict={list(set(housing_data["SellerG"]))[k]:k for k in range(len(set(housing_data["SellerG"])))}
RegionnameDict={list(set(housing_data["Regionname"]))[k]:k for k in range(len(set(housing_data["Regionname"])))}
CouncilAreaDict={list(set(housing_data["CouncilArea"]))[k]:k for k in range(len(set(housing_data["CouncilArea"])))}
PostcodeDict={list(set(housing_data["Postcode"]))[k]:k for k in range(len(set(housing_data["Postcode"])))}
housing_data["Suburb"]=housing_data["Suburb"].apply(lambda x:suburbDict[x])
housing_data["Type"]=housing_data["Type"].apply(lambda x:typeDict[x])
housing_data["Method"]=housing_data["Method"].apply(lambda x:methDict[x])
housing_data["SellerG"]=housing_data["SellerG"].apply(lambda x:sellerDict[x])
housing_data["Regionname"]=housing_data["Regionname"].apply(lambda x:RegionnameDict[x])
housing_data["CouncilArea"]=housing_data["CouncilArea"].apply(lambda x:CouncilAreaDict[x])
housing_data["Postcode"]=housing_data["Postcode"].apply(lambda x:PostcodeDict[x])
```

```python
YearBuiltAvg=housing_data[housing_data["YearBuilt"].notnull()]["YearBuilt"].mean()
avgBuildingAreaAvg=housing_data[housing_data["BuildingArea"].notnull()]["BuildingArea"].mean()
carAvg=housing_data[housing_data["Car"].notnull()]["Car"].mean()


housing_data=housing_data.fillna(value={"BuildingArea":avgBuildingAreaAvg,"YearBuilt":YearBuiltAvg,"Car":carAvg})
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
suburbClasses=len(suburbDict)
typeClasses=len(typeDict)
methodClasses=len(methDict)
sellerGClasses=len(sellerDict)
regionNameClasses=len(RegionnameDict)
councilAreaClasses=len(CouncilAreaDict)
postcodeClasses=len(PostcodeDict)

print(suburbClasses,typeClasses,methodClasses,sellerGClasses,regionNameClasses,councilAreaClasses,postcodeClasses)
```

```python
class NeuralNet(nn.Module):
    def __init__(self,):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(116, 256) 
        self.fc2 = nn.Linear(256, 256) 
        self.fc3 = nn.Linear(256, 1) 
        

        #I don't know how to use these embedding layers effectively but I've tried something here (this seems super unneeded cause only one lement is non-zero, this may not be what these are intended for)
         #number of embeddings for each

        suburbEmbeddings=15
        typeEmbeddings=5
        methodEmbeddings=5
        sellerEmbeddings=10
        regionnameEmbeddings=5
        councilAreaEmbeddings=5
        postcodeEmbeddings=10
        self.subEmb= nn.Embedding(suburbClasses, suburbEmbeddings)
        self.typeEmb= nn.Embedding(typeClasses, typeEmbeddings)
        self.methodEmb= nn.Embedding(methodClasses, methodEmbeddings)
        self.sellerEmb= nn.Embedding(sellerGClasses, sellerEmbeddings)
        self.regionnameEmb= nn.Embedding(regionNameClasses, regionnameEmbeddings)
        self.councilAreaEmb= nn.Embedding(councilAreaClasses, councilAreaEmbeddings)
        self.postcodeEmb= nn.Embedding(postcodeClasses, postcodeEmbeddings)

        
    def forward(self,numericalData, suburbVal,typeVal,methodVal,sellerVal,regionnameVal,councilAreaVal,postcodeVal):
        suburbEmbedded=self.subEmb(suburbVal)
        typeEmbedded=self.subEmb(typeVal)
        methodEmbedded=self.subEmb(methodVal)
        sellerEmbedded=self.subEmb(sellerVal)
        regionnameEmbedded=self.subEmb(regionnameVal)
        councilAreaEmbedded=self.subEmb(councilAreaVal)
        postcodeEmbedded=self.subEmb(postcodeVal)

        
        
        
        x=torch.cat((numericalData, suburbEmbedded,typeEmbedded,methodEmbedded,sellerEmbedded,regionnameEmbedded,councilAreaEmbedded,postcodeEmbedded),dim=1)
        x=self.fc1(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc2(x)
        x=F.relu(x)
        x=self.fc3(x)
        

        return x
```

```python
nn_trainset_predictions,nn_valset_predictions, nn_testset_predictions=[],[],[]


with torch.no_grad():
    #don't rearrange, just loop through (it may make more sense to just reset indexes for the sets when they are created, this is a bit confusing )
    X_train_temp=(X_trainsubset.copy(deep=True))
    y_train_temp=(y_trainsubset_label.copy(deep=True))
    X_val_temp=(X_valsubset.copy(deep=True))
    y_val_temp=(y_valsubset_label.copy(deep=True))
    X_test_temp=(X_testset.copy(deep=True))
    y_test_temp=(y_testset_label.copy(deep=True))


    X_train_temp=X_train_temp.reset_index()
    y_train_temp=y_train_temp.reset_index(drop=True)
    X_val_temp=X_val_temp.reset_index()
    y_val_temp=y_val_temp.reset_index(drop=True)
    X_test_temp=X_test_temp.reset_index()
    y_test_temp=y_test_temp.reset_index(drop=True)
    
    
    for index, row in X_train_temp.iterrows():
        input_data=X_train_temp.iloc[[index]]

        numerical_inputs=torch.from_numpy(np.array([[list(input_data["Rooms"])[0],
                                                    list(input_data["Distance"])[0],
                                                    list(input_data["Bedroom2"])[0],
                                                    list(input_data["Bathroom"])[0],
                                                    list(input_data["Car"])[0],
                                                    list(input_data["Landsize"])[0],
                                                    list(input_data["BuildingArea"])[0],
                                                    list(input_data["YearBuilt"])[0],
                                                    list(input_data["Lattitude"])[0],
                                                    list(input_data["Longtitude"])[0],
                                                    list(input_data["Propertycount"])[0],
                                                   ]]).astype(np.float32))
        output = model(numerical_inputs,
                      torch.LongTensor([int(list(input_data["Suburb"])[0])]),
                      torch.LongTensor([int(list(input_data["Type"])[0])]),
                      torch.LongTensor([int(list(input_data["Method"])[0])]),
                      torch.LongTensor([int(list(input_data["SellerG"])[0])]),
                      torch.LongTensor([int(list(input_data["Regionname"])[0])]),
                      torch.LongTensor([int(list(input_data["CouncilArea"])[0])]),
                      torch.LongTensor([int(list(input_data["Postcode"])[0])]),        
                    )
        nn_trainset_predictions.append(output[0])
    for index, row in X_val_temp.iterrows():
        input_data=X_val_temp.iloc[[index]]

        numerical_inputs=torch.from_numpy(np.array([[list(input_data["Rooms"])[0],
                                                    list(input_data["Distance"])[0],
                                                    list(input_data["Bedroom2"])[0],
                                                    list(input_data["Bathroom"])[0],
                                                    list(input_data["Car"])[0],
                                                    list(input_data["Landsize"])[0],
                                                    list(input_data["BuildingArea"])[0],
                                                    list(input_data["YearBuilt"])[0],
                                                    list(input_data["Lattitude"])[0],
                                                    list(input_data["Longtitude"])[0],
                                                    list(input_data["Propertycount"])[0],
                                                   ]]).astype(np.float32))
        output = model(numerical_inputs,
                      torch.LongTensor([int(list(input_data["Suburb"])[0])]),
                      torch.LongTensor([int(list(input_data["Type"])[0])]),
                      torch.LongTensor([int(list(input_data["Method"])[0])]),
                      torch.LongTensor([int(list(input_data["SellerG"])[0])]),
                      torch.LongTensor([int(list(input_data["Regionname"])[0])]),
                      torch.LongTensor([int(list(input_data["CouncilArea"])[0])]),
                      torch.LongTensor([int(list(input_data["Postcode"])[0])]),        
                    )
        nn_valset_predictions.append(output[0])
    
    for index, row in X_test_temp.iterrows():
        input_data=X_test_temp.iloc[[index]]

        numerical_inputs=torch.from_numpy(np.array([[list(input_data["Rooms"])[0],
                                                    list(input_data["Distance"])[0],
                                                    list(input_data["Bedroom2"])[0],
                                                    list(input_data["Bathroom"])[0],
                                                    list(input_data["Car"])[0],
                                                    list(input_data["Landsize"])[0],
                                                    list(input_data["BuildingArea"])[0],
                                                    list(input_data["YearBuilt"])[0],
                                                    list(input_data["Lattitude"])[0],
                                                    list(input_data["Longtitude"])[0],
                                                    list(input_data["Propertycount"])[0],
                                                   ]]).astype(np.float32))
        output = model(numerical_inputs,
                      torch.LongTensor([int(list(input_data["Suburb"])[0])]),
                      torch.LongTensor([int(list(input_data["Type"])[0])]),
                      torch.LongTensor([int(list(input_data["Method"])[0])]),
                      torch.LongTensor([int(list(input_data["SellerG"])[0])]),
                      torch.LongTensor([int(list(input_data["Regionname"])[0])]),
                      torch.LongTensor([int(list(input_data["CouncilArea"])[0])]),
                      torch.LongTensor([int(list(input_data["Postcode"])[0])]),        
                    )
        nn_testset_predictions.append(output[0])
```

```python
fig,ax=plt.subplots(1,1)
plt.scatter(list(y_trainsubset_label),nn_trainset_predictions,s=.01)
plt.scatter(list(y_valsubset_label),nn_valset_predictions,s=.01)
plt.scatter(list(y_testset_label),nn_testset_predictions,s=.01)
plt.xlabel("Actual Price")
plt.ylabel("NN Prediction")
ax.set_aspect("equal")

print("Train score: ",np.corrcoef(y_trainsubset_label, nn_trainset_predictions)[0,1]**2)
print("Val score: ",np.corrcoef(y_valsubset_label, nn_valset_predictions)[0,1]**2)
print("Test score: ",np.corrcoef(y_testset_label, nn_testset_predictions)[0,1]**2)

print("Train MSE: ",np.sqrt(np.mean(np.power((y_trainsubset_label-np.array(nn_trainset_predictions)),2))))
print("Val MSE: ",np.sqrt(np.mean(np.power((y_valsubset_label-np.array(nn_valset_predictions)),2))))
print("test MSE: ",np.sqrt(np.mean(np.power((y_testset_label-np.array(nn_testset_predictions)),2))))
```
